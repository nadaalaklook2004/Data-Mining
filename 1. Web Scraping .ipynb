```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Configigure Chrome options (to avoid detection)
options = webdriver.ChromeOptions()
options.add_argument("--disable-blink-features=AutomationControlled")   
options.add_experimental_option("excludeSwitches",["enable-automation"])
options.add_argument("--start-maximized")

#Initialize WebDriver
driver = webdriver.Chrome(service=Service(r"C:\Users\MSI\Desktop\NADA\nada ` s college\سنة ثانية الفصل الاول\التنقيب عن البيانات عملي\project\chromedriver-win64\chromedriver.exe"),options=options)




```

```python
import numpy as np
# load the page
required_pages = 24
myData = []
for page in range (required_pages) :
    url = f"https://www.carrefouruae.com/mafuae/ar/c/NF1220200?currentPage={page}"    
    driver.get(url)

    # Wait for listings to load
    try:
       WebDriverWait(driver, 5).until(
            EC.presence_of_element_located(By.CSS_SELECTOR,"article._63a946ba"))
    except:
        print("Listings didn't load properly")
    
    # parse HTML
    soup = BeautifulSoup(driver.page_source ,"html.parser")
    listings = soup.find_all("div",{"class":"relative w-[134px] max-w-[134px] sm:w-[215px] sm:max-w-[215px] flex justify-between"})


    for i in range(len(listings)):
        Extracted_data = {
                        "Price_after_discount":np.NaN,
                        "Price_before_discount": np.NaN,
                        "Remaining_quantity": np.NaN,
                        "Device_information": np.NaN ,
                        "Discount_rate" : np.NaN,
        }
        
        Price_after_discount = listings[i].find("div",{"class":"flex items-center my-xs"}).find("div",{"class":
            "text-lg leading-5 font-bold md:text-xl"})
        if Price_after_discount :
            Extracted_data['Price_after_discount'] = Price_after_discount.text.strip()
        
        Price_before_discount = listings[i].find("div",{"class":"flex items-center my-xs"}).find("div",{"class":
            "text-sm leading-4 font-medium text-gray-500 line-through"})
        if Price_before_discount :
            Extracted_data['Price_before_discount'] = Price_before_discount.text.strip()
            
        Remaining_quantity =  listings[i].find("div",{"class":"text-sm leading-4 font-regular text-shareprimary"})
        if Remaining_quantity :
            Extracted_data['Remaining_quantity'] = Remaining_quantity.text.strip()
            
        Device_information = listings[i].find("div",{"class":"text-sm leading-4 font-medium line-clamp-2 text-left md:text-md"})
        if Device_information :
            Extracted_data['Device_information'] = Device_information.text.strip()
            
        Discount_rate = listings[i].find("div",{"class":"inline-flex items-center justify-center font-bold rounded-bl-xs rounded-br-xl rounded-tl-xl rounded-tr-xs px-xs py-2xs text-[10px] bg-c4red-500 text-white flex max-h-[20px] relative max-w-[95%]"})
        if Discount_rate :        
            Extracted_data['Discount_rate'] = Discount_rate.text.strip()
            
        Extracted_data = {
                        "Price_after_discount":Price_after_discount,
                        "Price_before_discount": Price_before_discount,
                        "Remaining_quantity": Remaining_quantity,
                        "Device_information": Device_information,
                        "Discount_rate" : Discount_rate,
        }
        myData.append(Extracted_data)
```

```python
#The values in myData  are Tag from BeautifulSoup (like <div> or <span>)
#so we need code that processes each value if it is an HTML object and converts it to text using .get_text().
from bs4 import BeautifulSoup

for i in range(len(myData)):
    for key in myData[i]:
        value = myData[i][key]
        
        if hasattr(value, 'get_text'):  # إذا كان العنصر HTML
            myData[i][key] = value.get_text(" ", strip=True)
        else:
            myData[i][key] = str(value).strip() if value else None
```
